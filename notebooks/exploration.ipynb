# Exploration Postgres Transactions

Ce notebook illustre comment interroger la base Postgres de la pipeline pour des analyses ad hoc.


```python
import os

import pandas as pd
from sqlalchemy import create_engine  # type: ignore

POSTGRES_CONN_URI = os.getenv(
    "POSTGRES_CONN_URI",
    "postgresql+psycopg2://airflow:airflow@localhost:5432/transactions",
)
engine = create_engine(POSTGRES_CONN_URI, pool_pre_ping=True)

print("Connected to", POSTGRES_CONN_URI)
```

```python
latest_tx = pd.read_sql(
    """
    SELECT transaction_id, event_ts, user_id, amount, category, city, status
    FROM transactions_flat
    ORDER BY event_ts DESC
    LIMIT 20
    """,
    engine,
)
latest_tx
```

```python
daily_summary = pd.read_sql(
    """
    SELECT event_date, category, SUM(amount) AS total_amount, COUNT(*) AS tx_count
    FROM transactions_flat
    GROUP BY event_date, category
    ORDER BY event_date DESC, total_amount DESC
    LIMIT 50
    """,
    engine,
)
daily_summary
```

```python
engine.dispose()
```
